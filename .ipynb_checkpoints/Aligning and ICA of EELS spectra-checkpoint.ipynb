# Import DigitalMicrograph EELS files

```python
import hyperspy.api as hs

path = "mydata/myEELS/"

files = ["EELS Spectrum Image ZLP.dm3", 
         "EELS Spectrum Image CL1.dm3", 
         "EELS Spectrum Image CL2.dm3"] # ZLP and two core-loss signals

spectra = [] # Create list of SIs

for name in files:
    spectra.append(hs.load(path + name))

for SI in spectra:
    print(SI)
```

```python
# Align spectra according to ZLP 
spectra[0].align_zero_loss_peak(calibrate=True, also_align=spectra[1:]); 
# No negative signals
for s in spectra:
    s.data -= s.data.min() 
```

```python
# Set the notebook up for plotting using matplotlib
%matplotlib qt4 

spectra[2].plot() # Look at the plots 0,1,2
```

# Principal Component Analysis (PCA) Decomposition

```python
# Set plottling to inline mode, which are static but can be kept in the cell for later
%matplotlib inline

for s in spectra:
    
    s.decomposition(True) # 
    s.plot_explained_variance_ratio()
```

```python
eigenvalues = input("Number of eigenvalues counted from above (formatted like '6 6 3'): ")
eigenvalues = eigenvalues.split()
eigenvalues = [int(x) for x in eigenvalues]
```

# Independent Component Analysis (ICA)

```python
for (i, eigenvalue) in zip(range(len(spectra)), eigenvalues):
    spectra[i].blind_source_separation(eigenvalue)   
```

```python
# Change to interactive plotting
%matplotlib
%matplotlib qt4
# Manually iterate through spectra 0,1,2 to check the BSS results, go back and change eigenvalues if necessary
spectra[2].plot_bss_results()
```

```python
bss_spectra = []
for s in spectra:
    bss_spectra.append(s.get_bss_model())

for SI in bss_spectra:
    print(SI)
```

# Save spectra for hyperspy and DM

```python
for (sc, name) in zip(bss_spectra, files):
    sc.save(path + "bss/" + name + " bss.hdf5") # For Hyperspy
    sc.save(path + "bss/" + name + " bss.rpl") # For DigitalMicrograph import, using the importRPL script
```
